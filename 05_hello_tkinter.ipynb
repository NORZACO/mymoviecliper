```python
# pylint: disable=missing-final-newline
#!/usr/bin/python3
# widgets.py by Barron Stone
# This is an exercise file from Python GUI Development with Tkinter on lynda.com
from tkinter import *
from tkinter import ttk
class VideoEditorGUI:
    def __init__(self, root):
        self.root = root
        self.root.geometry("500x500")
        self.root.title("Video Editor")
        self.setup_gui()
        
        
    def setup_gui(self):
        test_button = ttk.Button(self.root, text="Select Images", width=25, command=self.testingCommands)
        test_button.pack(pady=10)
        
        print(test_button["text"])
        test_button["text"] = "Test Code"
        test_button.config(text = "Testing Tk Code")
        
        # print(test_button.config())
        print(str(test_button), self.root)
        ttk.Button(self.root, text="STOP", width=25, command=self.root.destroy).pack(pady=10)
        ttk.Button(self.root, text="GeometryCommands", width=25, command=self.GeometryCommands).pack(pady=10)
        ttk.Button(self.root, text="Labeling", width=25, command=self.Labeling).pack(pady=10)
    
    
    def testingCommands(self):
        print("Testing Commands")
        ttk.Label(self.root, text= "Hello Thinker", background="green").pack(pady=10)
        
    
    def Labeling(self):
        print("GeometryCommands")
        ttk.Label(self.root, text= "GUI DEV", background="purple").pack(pady=10)
        
    def GeometryCommands(self):
        print("GeometryCommands")

        

        

    
    
if __name__ == "__main__":
    root = Tk()
    app = VideoEditorGUI(root)
    root.mainloop()
```

```python
# pylint: disable=missing-final-newline
#!/usr/bin/python3
# widgets.py by Barron Stone
# This is an exercise file from Python GUI Development with Tkinter on lynda.com



from tkinter import *
from tkinter import ttk
class HelloApp:
    def __init__(self, master):
        self.label = ttk.Label(master, text = "Hello World", justify=CENTER)
        self.label.grid(row = 0, column = 0, columnspan = 2)
        
        ttk.Button(master, text = "STOCKHOLM", command = self.denmark_hello).grid(row = 2, column = 0)
        ttk.Button(master, text = "OSLO S", command = self.norge_hello).grid(row = 2, column = 2)
        
    def denmark_hello(self):
       self.label.config(text = 'Im from DenMark, I love Tkinter!', background="green", justify=CENTER, wraplength=150)

    def norge_hello(self):
       self.label.config(font="Courier", text = 'Im from Norway, I do coding in Tkinter!', background="yellow", justify=CENTER,wraplength=150)
       

    
    
if __name__ == "__main__":
    root = Tk()
    root.title("Video Editor for TikTok")
    root.geometry("500x200")
    root.configure(bg="white")  # Set background color to white
    app = HelloApp(root)
    root.mainloop()
```
